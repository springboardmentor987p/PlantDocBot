```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
encoder = LabelEncoder()
encoder.fit(df["caption"].tolist())
df["label"] = encoder.transform(df["caption"].tolist())
df
```

```python
df_train, df_test = train_test_split(df, train_size=0.8)
```

```python
from datasets import Dataset
```

```python
train_dataset = Dataset.from_pandas(df_train)
test_dataset = Dataset.from_pandas(df_test)
```

```python
from transformers import AutoTokenizer
```

```python
model_name = "distilbert-base-uncased"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
def mapper_function(data):
  return tokenizer(data['captions'],truncation=True)
```

```python
tokenized_train = train_dataset.map(mapper_function, batched = True)
tokenized_train = test_dataset.map(mapper_function, batched = True)
```

```python
from transformers import AutoModelForSequenceClassification
```

```python
model = AutoModelForSequenceClassification.from_pretrained(model_name,num_labels=15)
```

```python
from transformers import Trainer, TrainingArguments
from transformers import DataCollator
import evaluate
import numpy as np
```

```python
from transformers import DataCollatorWithPadding

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

```python
eval_metric = evaluate.load("accuracy")
```

```python
def compute_metrics(pred):
  logits,labels = pred
  prediction = np.argmax(logits,axis=1)
```

```python
from transformers import Trainer, TrainingArguments

training_arguments = TrainingArguments(
    output_dir="checkpoints",
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    learning_rate=1e-4,
    num_train_epochs=5,  # corrected
    weight_decay=0.01,
    logging_strategy="epoch",
    save_strategy="epoch",
    save_total_limit=2,
    report_to="none",
)


```

```python
# 1. Define a tokenization function
def tokenize_function(examples):
    return tokenizer(examples["caption"], padding="max_length", truncation=True)

# 2. Apply it to train and test datasets
tokenized_train = train_dataset.map(tokenize_function, batched=True)
tokenized_test = test_dataset.map(tokenize_function, batched=True)
```

```python
trainer = Trainer(
    model=model,
    args=training_arguments,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_test,
    tokenizer=tokenizer,
    data_collator=data_collator,
)
```

```python
trainer.train()
```
