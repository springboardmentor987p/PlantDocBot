```python
from transformers import AutoTokenizer
```

```python
model_name = "distilbert-base-uncased"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
def mapper_function(data):
  return tokenizer(data['captions'],truncation=True)
```

```python
tokenized_train = train_dataset.map(mapper_function, batched = True)
tokenized_train = test_dataset.map(mapper_function, batched = True)
```

```python
from transformers import AutoModelForSequenceClassification
```

```python
model = AutoModelForSequenceClassification.from_pretrained(model_name,num_labels=15)
```

```python
from transformers import Trainer, TrainingArguments
from transformers import DataCollator
import evaluate
import numpy as np
```

```python
from transformers import DataCollatorWithPadding

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

```python
eval_metric = evaluate.load("accuracy")
```

```python
def compute_metrics(pred):
  logits,labels = pred
  prediction = np.argmax(logits,axis=1)
  return eval_metric.compute(prediction,labels)
```

```python
from transformers import Trainer, TrainingArguments

training_arguments = TrainingArguments(
    output_dir="checkpoints",
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    learning_rate=1e-4,
    num_train_epochs=3,  # corrected
    weight_decay=0.01,
    logging_strategy="epoch",
    save_strategy="epoch",
    save_total_limit=2,
    report_to="none",
)


```

```python
# 1. Define a tokenization function
def tokenize_function(examples):
    return tokenizer(examples["caption"], padding="max_length", truncation=True)

# 2. Apply it to train and test datasets
tokenized_train = train_dataset.map(tokenize_function, batched=True)
tokenized_test = test_dataset.map(tokenize_function, batched=True)
```

```python
trainer = Trainer(
    model=model,
    args=training_arguments,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_test,
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics
)
```

```python
trainer.train()
```

```python
# Save the model and tokenizer to a folder called 'saved_model'
trainer.save_model("saved_model")
tokenizer.save_pretrained("saved_model")
```

```python
import pandas as pd
import torch
from transformers import AutoModelForSequenceClassification, AutoTokenizer


df = pd.read_parquet("hf://datasets/ButterChicken98/plantvillage-image-text-pairs/data/train-00000-of-00001.parquet")

print(df.columns)
print(df.head())
class_column_name = 'caption'
class_names = df[class_column_name].unique().tolist()
print("Class names extracted:")
print(class_names)
model_path = "saved_model"
model = AutoModelForSequenceClassification.from_pretrained(model_path)
tokenizer = AutoTokenizer.from_pretrained(model_path)
def predict(text):
    model.eval()
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)
    with torch.no_grad():
        outputs = model(**inputs)
        predicted_idx = torch.argmax(outputs.logits, dim=1).item()
    return class_names[predicted_idx]
sample_text = "image of a healthy plant leaf"
prediction = predict(sample_text)
print(f"Predicted class: {prediction}")
```
