```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
encoder = LabelEncoder()
encoder.fit(df["caption"].tolist())
df["label"] = encoder.transform(df["caption"].tolist())
df
```

```python
df_train, df_test = train_test_split(df, train_size=0.8)
```

```python
from datasets import Dataset
train_dataset = Dataset.from_pandas(df_train)
test_dataset = Dataset.from_pandas(df_test)
```

```python
from transformers import AutoTokenizer
```

```python
model_name = "distilbert-base-uncased"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
def mapper_function(data):
  return tokenizer(data['captions'], truncation=True)
```

```python
tokenized_train = train_dataset.map(mapper_function, batched=True)
tokenized_test = test_dataset.map(mapper_function, batched=True)
```

```python
from transformers import AutoModelForSequenceClassification
```

```python
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=15)
```

```python
!pip install evaluate
```

```python
from transformers import Trainer, TrainingArguments
from transformers import DataCollatorWithPadding
import evaluate
import numpy as np
```

```python
data_collator = DataCollatorWithPadding(tokenizer)
```

```python
eval_metric = evaluate.load("accuracy")
```

```python
def compute_metrics(eval_pred):
  logits, labels = eval_pred
  preds = np.argmax(logits, axis=1)
  return eval_metric.compute(predictions=preds, references=labels)
```

```python
training_arguments = TrainingArguments(
    output_dir="checkpoints",
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    learning_rate=1e-4,
    num_train_epochs=5,
    weight_decay=0.01,
    logging_strategy="epoch",
    save_strategy="epoch",
    save_total_limit=5,
    report_to="none"
    )
```

```python
trainer = Trainer(
    model=model,
    args=training_arguments,
    train_dataset = tokenized_train,
    eval_dataset = tokenized_test,
    tokenizer= tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics
    )
```

```python
trainer.train()
```

**a) Using Hugging Face pipeline (easy)**

```python
classifier = pipeline("text-classification", model=model_loaded, tokenizer=tokenizer_loaded, return_all_scores=False)

sample_texts = [
    "leaf with brown spots and yellowing around the edges",
    "healthy green leaf with no spots"
]
pipe_out = classifier(sample_texts, truncation=True)
```

```python
pred_ids_from_pipeline = [int(d["label"].split("_")[-1]) for d in pipe_out]
pred_labels_from_pipeline = encoder_loaded.inverse_transform(pred_ids_from_pipeline)
print("Pipeline predictions:", list(zip(sample_texts, pred_labels_from_pipeline)))
```

**b) Raw PyTorch inference (fine-grained control)**

```python
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model_loaded.to(device)    # move model to GPU if available
model_loaded.eval()

# move inputs to the same device
inputs = tokenizer_loaded(sample_texts, truncation=True, padding=True, return_tensors="pt")
inputs = {k: v.to(device) for k, v in inputs.items()}

with torch.no_grad():
    outputs = model_loaded(**inputs)
    logits = outputs.logits
    preds = torch.argmax(logits, dim=-1).cpu().numpy()  # move back to CPU for numpy

decoded = encoder_loaded.inverse_transform(preds)
print("Raw model predictions:", list(zip(sample_texts, decoded)))
```

```python
import torch
import joblib
from transformers import AutoModelForSequenceClassification, AutoTokenizer, pipeline

def load_saved_model(model_path="checkpoints/final_model"):
    """
    Loads the trained model, tokenizer, and label encoder from disk.
    Returns: model, tokenizer, encoder, device
    """
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    model = AutoModelForSequenceClassification.from_pretrained(model_path)
    model.to(device)
    model.eval()

    tokenizer = AutoTokenizer.from_pretrained(model_path)
    encoder = joblib.load(f"{model_path}/label_encoder.joblib")

    return model, tokenizer, encoder, device


def predict_text(texts, model, tokenizer, encoder, device):
    """
    Predict the class for a list of texts.
    texts: list of strings
    Returns: list of predicted class labels
    """
    # Tokenize and move to correct device
    inputs = tokenizer(texts, truncation=True, padding=True, return_tensors="pt")
    inputs = {k: v.to(device) for k, v in inputs.items()}

    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits
        pred_ids = torch.argmax(logits, dim=-1).cpu().numpy()

    pred_labels = encoder.inverse_transform(pred_ids)
    return pred_labels


# ----------------- USAGE -----------------

# Load model once
model, tokenizer, encoder, device = load_saved_model("checkpoints/final_model")

# Predict on new text
sample_texts = [
    "leaf with brown spots and yellowing around the edges",
    "healthy green leaf with no spots"
]

predictions = predict_text(sample_texts, model, tokenizer, encoder, device)
for text, label in zip(sample_texts, predictions):
    print(f"Text: '{text}' --> Predicted label: {label}")
```
